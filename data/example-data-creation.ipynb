# Creating Useful Example Data

This notebook explores different ways of creating useful dummy data. 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

## Linear

```python
x = np.linspace(0, 1, 2)
y = np.linspace(0, 1, 2)
plt.figure(figsize=(5, 5))
plt.plot(x, y)
plt.show()
```

## Periodic

```python
x = np.linspace(0, 12, 365)
y = ((1 + np.sin((x - 5.0) / 2.0)) * 15) 
y += np.sin(y) * np.random.ranf(365) * 3 
plt.figure(figsize=(20, 5))
plt.plot(x, y)
plt.title('Daily Tempuratures')
plt.xlabel('Month')
plt.ylabel('Celsius')
plt.show()
```

## Clusters

```python
# make random variables repeatable
np.random.seed(1981)

# create clusters of nodes 
# params - (mean, covariance matrix, number of observations)
c1 = np.random.multivariate_normal([5, 22], [[0.0, 1], [1, 0]], size=100)
c2 = np.random.multivariate_normal([8, 10], [[2, 0.15], [0.15, 2]], size=15)
c3 = np.random.multivariate_normal([20, 5], [[1, 0.5], [0.5, 1]], size=50)
c4 = np.random.multivariate_normal([22, 25], [[1, 1.0], [1.0, 5]], size=100)
c5 = np.random.multivariate_normal([30, 20], [[1, 0.2], [0.2, 3]], size=20)

# plot it
plt.figure(figsize=(5, 5))
plt.scatter(c1[:,0], c1[:,1], color='c')
plt.scatter(c2[:,0], c2[:,1], color='m')
plt.scatter(c3[:,0], c3[:,1], color='r')
plt.scatter(c4[:,0], c4[:,1], color='orange')
plt.scatter(c5[:,0], c5[:,1], color='b')
plt.show()
```
