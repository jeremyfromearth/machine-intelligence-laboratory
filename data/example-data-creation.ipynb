# Creating Example Data

This notebook explores different ways of creating useful dummy data. 

```python
import numpy as np
import pandas as pd
from scipy.misc import factorial 
from matplotlib.pyplot import *
%matplotlib inline
```

## Linear

```python
x = np.linspace(0, 1, 2)
y = np.linspace(0, 1, 2)
figure(figsize=(5, 5))
grid(True)
plot(x, y)
```

## Quadratic

```python
x = np.linspace(0, 1, 100)
y = x ** 2
figure(figsize=(5, 5))
grid(True)
plot(x, y)
```

## Polynomial

```python
x = np.linspace(-5, 5, 200)
y = (x**2 + x**3 + x**4 + x**5) * np.exp(-x ** 2)
figure(figsize=(5, 5))
grid(True)
plot(x, y)
```

## Poisson Distribution

```python
r = 15.0    # rate / average 
k = r * 2.0
x = np.linspace(0, k, k + 1)
p = ((r ** x) * np.exp(-r)) / factorial(x)
plot(x, p)
scatter(x, p)
```

## Periodic

```python
# from 
x = np.linspace(0, 12, 365)
y = (1 + np.sin((x - 4) / (12 / (2 * np.pi)))) * 15
y += np.sin(y) * np.random.ranf(365) * 3 
figure(figsize=(20, 5))
grid(True)
plot(x, y)
title('Daily Tempuratures')
xlabel('Month')
ylabel('Celsius')
```

## Clusters

```python
# make random variables repeatable
np.random.seed(1981)

# create clusters of nodes 
# params - (mean, covariance matrix, number of observations)
c1 = np.random.multivariate_normal([5, 22], [[0.0, 1], [1, 0]], size=100)
c2 = np.random.multivariate_normal([8, 10], [[2, 0.15], [0.15, 2]], size=15)
c3 = np.random.multivariate_normal([20, 5], [[1, 0.5], [0.5, 1]], size=50)
c4 = np.random.multivariate_normal([22, 25], [[1, 1.0], [1.0, 5]], size=100)
c5 = np.random.multivariate_normal([30, 20], [[1, 0.2], [0.2, 3]], size=20)

# plot it
figure(figsize=(5, 5))
scatter(c1[:,0], c1[:,1], color='c')
scatter(c2[:,0], c2[:,1], color='m')
scatter(c3[:,0], c3[:,1], color='r')
scatter(c4[:,0], c4[:,1], color='orange')
scatter(c5[:,0], c5[:,1], color='b')
```
