# Creating Useful Example Data

This notebook explores different ways of creating useful dummy data. 

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## Clusters

```python
# make random variables repeatable
np.random.seed(1981)

# create clusters of nodes 
# params - (mean, covariance matrix, number of observations)
c1 = np.random.multivariate_normal([15, 20], [[0.0, 1], [1, 0]], size=100)
c2 = np.random.multivariate_normal([15, 10], [[2, 0.15], [0.15, 2]], size=15)
c3 = np.random.multivariate_normal([20, 5], [[1, 0.5], [0.5, 1]], size=50)
c4 = np.random.multivariate_normal([20, 25], [[1, 1.0], [1.0, 10]], size=100)
c5 = np.random.multivariate_normal([30, 20], [[1, 0.2], [0.2, 4]], size=20)

# plot it
plt.figure(figsize=(10, 10))
axes = plt.gca()
axes.set_xlim([0,40])
axes.set_ylim([0,40])
plt.scatter(c1[:,0], c1[:,1], color='c')
plt.scatter(c2[:,0], c2[:,1], color='m')
plt.scatter(c3[:,0], c3[:,1], color='r')
plt.scatter(c4[:,0], c4[:,1], color='orange')
plt.scatter(c5[:,0], c5[:,1], color='b')
plt.show()
```
