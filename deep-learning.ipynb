# Prepare Non-linearly Separable Data

```python
Y = np.array([])
X1 = np.array([])
X2 = np.array([])
classes = ['.', '^', 'v', 's', '8']

start = 0
n_classes = len(classes)
inc = (np.pi * 2) / n_classes
points_per_class = 256
figure(figsize=(10, 10))
for i in range(0, n_classes):
    end = start + inc + np.random.uniform(0, 0.2)
    r = np.random.uniform(start, end, points_per_class)
    x1 = np.sin(r) * np.random.uniform(1, 1.5, points_per_class)
    x2 = np.cos(r) * np.random.uniform(1, 1.5, points_per_class)
    start += inc
    scatter(x1, x2, marker=classes[i])
    
    X1 = np.concatenate((X1, x1), axis=0)
    X2 = np.concatenate((X2, x2), axis=0)
    Y = np.concatenate((Y, [i for x in range(0, points_per_class)]), axis=0)
```
